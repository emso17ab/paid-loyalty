```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
import matplotlib.pyplot as plt

# Listing all the variables
program_attractiveness = 0
switching_costs = 0
market_demand = 0
quality_perception = 0

income = 0
economic_orientation = 0

purchase_freq = 0
revenue = 0
sow = 0
satisfaction = 0
paid_membership = 0

NROWS = 10000

df = pd.DataFrame(np.random.normal(size=(NROWS,4)), columns=['program_attractiveness', 'switching_costs', 'market_demand', 'quality_perception'])

df['income'] = np.random.randint(15000, 100000, NROWS)

df['economic_orientation'] = np.random.normal(size=NROWS)
df['economic_orientation'] = df['economic_orientation'] + df['income'].apply(math.log)


df['paid_membership'] = np.random.normal(scale=3.0, size=NROWS)
w1, w2, w3, w4, w5 = 1.5, -1.4, 1.8, 0.00004, 0.
df['paid_membership'] = df['paid_membership'] + df['program_attractiveness']*w1 + df['switching_costs']*w2 + df['quality_perception']*w3 + df['income']*w4 + df['economic_orientation']*w5
df['paid_membership'] = df['paid_membership'] > 4


df['satisfaction'] = np.random.uniform(df['paid_membership'].astype(int), 3.0 + df['paid_membership'].astype(int), size=NROWS) + np.random.lognormal(mean=2.5, sigma=0.5, size=NROWS)
df['satisfaction'] = df['satisfaction'] + df['quality_perception']


```

```python
sns.histplot(df.satisfaction)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x="quality_perception", y="satisfaction", hue='paid_membership')
```
